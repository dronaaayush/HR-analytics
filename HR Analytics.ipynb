```python
import pandas as pd
df = pd.read_csv('HR_Analytics.csv')
df
```

```python
#3. Attrition Rate by Demographics (Environment Satisfaction)
import pandas as pd
import matplotlib.pyplot as plt

attrition_rate = df[df['Attrition']=='Yes']
total_by_satisf = df['EnvironmentSatisfaction'].value_counts().sort_index()
attrition_rate_by_satisf = attrition_rate['EnvironmentSatisfaction'].value_counts().sort_index()
attrition_actual_rate = (attrition_rate_by_satisf/total_by_satisf)*100

attrition_actual_rate.plot(kind='bar', color = 'blue', figsize= (9, 9))
plt.title('Attrition Rate By Employee Satisfaction')
plt.xlabel('Environment Satisfaction')
plt.ylabel('Attrition Rate')
plt.show()

```

```python
# Salary Distribution Analysis  (By Job Level)

job_level_salary = df.groupby('JobLevel')['MonthlyIncome'].describe()

job_level_salary.plot(figsize=(9, 9), kind='bar' )
plt.show()
```

```python
# Salary Distribution Analysis  (By Department)

depart_salary = df.groupby('Department')['MonthlyIncome'].describe()
depart_salary.plot(kind='bar', figsize=(9, 9),)
plt.show()
```

```python
# Salary Distribution Analysis  (Correlation between PercentSalaryHike and PerformanceRating)

correlation = df[['PercentSalaryHike', 'PerformanceRating']].corr()
print(correlation)

correlation.plot(kind='bar', x='PerformanceRating', y='PercentSalaryHike')
plt.title('Salary Hikes vs. Performance Rating')
plt.show()
```

```python
# Group data by YearsSinceLastPromotion and calculate average satisfaction

import seaborn as sns

promotion_satisfaction = df.groupby('YearsSinceLastPromotion')['JobSatisfaction'].mean()
print(promotion_satisfaction)

sns.barplot(data=df, x='YearsSinceLastPromotion', y='JobSatisfaction', palette='viridis')
plt.title('Job Satisfaction vs. Time Since Last Promotion')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Average Job Satisfaction')
plt.show()
```
